```python
#parent (super) class
class animals:
    def __init__(self, name):
        self.name = name
        print(f"The name of the animal is {self.name}")
        
    def char(self):
        print("Animals are multicellular organisms")
        
ani_1 = animals("Furry")
ani_1.char()
print("\n")
        

#single level and hierarchical inheritance
class herbivores(animals):
    def __init__(self, name):
        self.name = name
        print(f"{self.name} is a herbivorous animal")
        
    def eat(self):
        print("They eat grass")
        
ani_2 = herbivores("Cow")
ani_2.char()
ani_2.eat()
print("\n")
        
class carnivores(animals):
    def __init__(self, name):
        self.name = name
        print(f"{self.name} is a carnivorous animal")
        
    def eat(self):
        print("They eat meat")
        
ani_3 = carnivores("Lion")
ani_3.char()
ani_3.eat()
print("\n")
        
        
#multilevel inheritance
class tiger(carnivores):
    def __init__(self):
        print("Tigers are one of the largest cats in the world")
    
    def body(self):
        print("Tigers have reddish-orange coats with black stripes")
        
ani_4 = tiger()
ani_4.char()
ani_4.eat()
ani_4.body()
print("\n")
        
class rabbit(herbivores):
    def __init__(self):
        print("Rabbits have long sharp incisors")
    
    def body(self):
        print("Rabbits are grey or brown and have long ears and short tails")
        
ani_5 = rabbit()
ani_5.char()
ani_5.eat()
ani_5.body()
print("\n")
        
        
#multiple inheritance
class dog(herbivores, carnivores):
    def __init__(self):
        print("Dogs are herbivores as well as carnivores i.e. they are omnivores")
        
    def body(self):
        print("Dogs have a coat of hair and sharp teeth")
        
ani_6 = dog()
ani_6.eat()
ani_6.body()
```
